#### **5.5. Machine Learning Models**

- **Logistic Regression (LR)**:
  Logistic regression is a classification algorithm that models the probability of a binary outcome based on one or more predictor variables using a logistic function.
  
- **Support Vector Machines (SVM)**:
  SVM finds a hyperplane that best separates the data into classes by maximizing the margin between the closest points of the classes (support vectors).

- **Stochastic Gradient Descent (SGD)**:
  An iterative optimization algorithm used for minimizing loss functions in linear classifiers and regressors, particularly useful for large datasets.

- **Decision Trees**:
  A tree-based algorithm where each internal node represents a decision on a feature, each leaf node represents an outcome, and paths from root to leaf represent classification rules.

- **K-Nearest Neighbors (KNN)**:
  A lazy learning algorithm that classifies a data point based on the majority class of its k-nearest neighbors in the feature space.
  
- **Naive Bayes (NB)**:
  A probabilistic classifier based on Bayes' theorem, assuming strong independence between features.

- **Bagging (Bootstrap Aggregating)**:
  Combines predictions from multiple models (e.g., Random Forest) to reduce variance and improve accuracy.

- **Boosting**:
  Combines weak learners sequentially to form a strong learner by minimizing errors iteratively (e.g., Extreme Gradient Boosting - XGBoost).

#### **5.6. Model Performance Metrics**

- **Confusion Matrix**:
  - Sensitivity (Recall): True Positive Rate
  - Specificity: True Negative Rate
  - Accuracy: Overall correctness
  - Precision: Positive Predictive Value
  - F1-Score: Harmonic mean of Precision and Recall
  - AUC: Area Under the ROC Curve

- **K-Fold Cross-Validation**: Splits data into k subsets; trains on k-1 and tests on the remaining one iteratively to ensure robust performance evaluation.

- **Model Run Statistics**:
  - Time Taken
  - Memory Used
  - Inherent Complexity

###**7.2 Insights**

### **Logistic Regression (LR)**
#### **Key Features and Thresholds:**
- **Key Features:**
  - Focus on **highly correlated features** (e.g., categorical variables or numeric predictors with strong linear relationships) for better interpretability and model clarity.
  - Evaluate the impact of each feature through **coefficients**, which indicate the relationship strength and direction (positive or negative) with the outcome.
  
- **Threshold:**
  - Use **multiple decision thresholds** (e.g., 0.5, 0.6) to find a balance between **precision and recall**, especially in underrepresented classes. For example, a lower threshold can increase recall but at the cost of precision, which is critical for imbalanced datasets.

#### **Insights on Uniqueness, Usability, and Suitability:**
- **Uniqueness:** Logistic Regression is highly **simple and interpretable**, making it ideal for datasets where classes are linearly separable.
- **Usability:** It is best used for **quick preliminary analysis** when time and interpretability are crucial, allowing decision-makers to easily understand the effects of variables.
- **Suitability:** This model **performs poorly** when the data is non-linear or has severe **class imbalance**. It is more effective with **balanced** and **linearly distributed** datasets, making it limited in more complex scenarios.

---

### **Support Vector Machine (SVM)**
#### **Key Features and Thresholds:**
- **Key Features:**
  - Focus on features that **enhance separability** between classes, leveraging high-dimensional spaces. It is crucial to select features that maximize the margin between classes, especially in non-linear separations.
  
- **Threshold:**
  - Use a **soft margin** approach, where you adjust **hyperparameters** (like **C** and **gamma**) to balance **precision** and **recall** for underrepresented or minority classes. This helps in avoiding overfitting, especially in imbalanced datasets.

#### **Insights on Uniqueness, Usability, and Suitability:**
- **Uniqueness:** SVM is excellent at **capturing complex decision boundaries**, particularly with **kernel tricks** that can handle non-linear relationships in high-dimensional data spaces.
- **Usability:** SVM is often less **interpretable** and more **computationally expensive**, making it suitable for **small, high-dimensional datasets**. The training process can be time-consuming depending on the kernel used.
- **Suitability:** It **performs poorly on imbalanced datasets** due to its sensitivity to class distributions. However, it is effective for datasets where **capturing subtle patterns** is essential, such as image recognition or text classification.

---

### **Decision Tree (DT)**
#### **Key Features and Thresholds:**
- **Key Features:**
  - Features with high **information gain** or **Gini index** should be selected to split nodes, as they provide the most significant contribution to classifying the data.
  
- **Threshold:**
  - **Adjust tree depth** to avoid overfitting and ensure interpretability. A **max depth of 5** is typically a good threshold to balance between model complexity and performance, especially in small datasets.

#### **Insights on Uniqueness, Usability, and Suitability:**
- **Uniqueness:** Decision Trees provide **interpretable, tree-structured decisions**, making it easier to visualize the logic behind predictions and capture **non-linear relationships** in the data.
- **Usability:** It is easy to **visualize** and **explain**, providing clear decision rules for stakeholders, making it an excellent tool for decision-making.
- **Suitability:** Decision Trees are **effective for moderately complex datasets**, but they are prone to **overfitting** in small datasets. They work well with **balanced** and moderately imbalanced data but may struggle with noise.

---

### **Random Forest (RF)**
#### **Key Features and Thresholds:**
- **Key Features:**
  - Use **ensemble averaging** to improve predictive accuracy and reduce variance across multiple decision trees. This enhances model stability.
  
- **Threshold:**
  - Focus on selecting the **optimal number of trees** (e.g., **50-100**) and **features per split** to balance performance and resource consumption. Too many trees can increase computational cost.

#### **Insights on Uniqueness, Usability, and Suitability:**
- **Uniqueness:** Random Forest combines multiple decision trees to improve **stability** and **generalization**, significantly reducing overfitting compared to a single decision tree.
- **Usability:** It is **computationally more expensive** than a single decision tree, requiring more memory and time. However, it performs better in handling **noisy** and **imbalanced data**.
- **Suitability:** Random Forest is ideal for **larger datasets** with **complex relationships**, where interpretability is less critical than **predictive accuracy**. It works well in both balanced and imbalanced datasets.

```python
# Confidence Interval for the Mean
def confidence_interval(data, confidence=0.95):
    mean = data.mean()
    std = data.std()
    n = len(data)
    z = stats.norm.ppf(1 - (1 - confidence) / 2)  # z-score for 95% CI
    margin_of_error = z * (std / np.sqrt(n))
    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error
    return (lower_bound, upper_bound)

ci_results = {}
for col in sr41_non_cat_vars:
    ci_results[col] = confidence_interval(sr41_sample[col])


print("\nConfidence Intervals (95%):")
for col, ci in ci_results.items():
    print(f"{col}: {ci[0]:.2f} to {ci[1]:.2f}")
```

## **Inferential Statistics**

```python
# Test of Homogeneity {Chi-sq} on categorical data

sr41_contingency_table = pd.crosstab(sr41_sample['Import_Export'], sr41_sample['Category'])

# Perform the Chi-square test
chi2, p, dof, expected = chi2_contingency(sr41_contingency_table)

print(f"Chi-square statistic: {chi2}")
print(f"P-value: {p}")
print(f"Degrees of freedom: {dof}")
print("Expected frequencies:")
print(expected)

# Interpret the results
sr41_alpha = 0.05  # Significance level

if p < sr41_alpha:
    print("Reject the null hypothesis: There is a significant association between the two categorical variables.")
else:
    print("There is no significant association between the two categorical variables.")
```

```python
# Test of Normality {Shapiro-Wilk | Kolmogorov-Smirnov | Anderson-Darling | Jarque-Bera

def normality_tests(data):

    sr41_results = {}

    # Shapiro-Wilk test
    shapiro_statistic, shapiro_pvalue = shapiro(data)
    sr41_results["Shapiro-Wilk"] = {"statistic": shapiro_statistic, "p-value": shapiro_pvalue}

    # Kolmogorov-Smirnov test (against a normal distribution)
    ks_statistic, ks_pvalue = kstest(data, 'norm')
    sr41_results["Kolmogorov-Smirnov"] = {"statistic": ks_statistic, "p-value": ks_pvalue}

    # Anderson-Darling test
    anderson_result = anderson(data)
    sr41_results["Anderson-Darling"] = {
        "statistic": anderson_result.statistic,
        "critical_values": anderson_result.critical_values,
        "significance_level": anderson_result.significance_level
    }

    # Jarque-Bera test
    jb_statistic, jb_pvalue = jarque_bera(data)
    sr41_results["Jarque-Bera"] = {"statistic": jb_statistic, "p-value": jb_pvalue}

    return sr41_results
```

```python
from scipy import stats

# Test of Correlation (T-test)
def t_test_correlation(data1, data2):
    t_statistic, p_value = stats.ttest_ind(data1, data2)
    return {"t-statistic": t_statistic, "p-value": p_value}

for col in sr41_non_cat_vars:
    print(f"Normality tests for {col}:")
    normality_test_results = normality_tests(sr41_sample[col])
    for test_name, result in normality_test_results.items():
        print(f"  {test_name}: {result}")

    # Example of T-test
    if len(sr41_non_cat_vars) > 1:
        for i in range(len(sr41_non_cat_vars)):
            for j in range(i+1, len(sr41_non_cat_vars)):
                col1, col2 = sr41_non_cat_vars[i], sr41_non_cat_vars[j]
                t_test_result = t_test_correlation(sr41_sample[col1], sr41_sample[col2])
                print(f"T-test Correlation between {col1} and {col2}: {t_test_result}")
    print()
    print('*'*20)
    print()
```

##Time Taken


```python
import time
from sklearn.linear_model import LogisticRegression, SGDClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier

def time_model_run(model, X_train, y_train):
    start_time = time.time()
    model.fit(X_train, y_train)
    end_time = time.time()
    return end_time - start_time

# Logistic Regression
log_reg_model = LogisticRegression(random_state=55041)
log_reg_time = time_model_run(log_reg_model, sr41_X_train, sr41_y_train)
print(f"Logistic Regression runtime: {log_reg_time:.4f} seconds")

# Support Vector Machine (SVM)
svm_model = SVC(random_state=55041)
svm_time = time_model_run(svm_model, sr41_X_train, sr41_y_train)
print(f"SVM runtime: {svm_time:.4f} seconds")

# Stochastic Gradient Descent (SGD)
sgd_model = SGDClassifier(random_state=55041)
sgd_time = time_model_run(sgd_model, sr41_X_train, sr41_y_train)
print(f"SGD runtime: {sgd_time:.4f} seconds")

# Decision Tree
dt_model = DecisionTreeClassifier(random_state=55041)
dt_time = time_model_run(dt_model, sr41_X_train, sr41_y_train)
print(f"Decision Tree runtime: {dt_time:.4f} seconds")

# K-Nearest Neighbors (KNN)
knn_model = KNeighborsClassifier(n_neighbors=5)
knn_time = time_model_run(knn_model, sr41_X_train, sr41_y_train)
print(f"KNN runtime: {knn_time:.4f} seconds")

# Naive Bayes
nb_model = GaussianNB()
nb_time = time_model_run(nb_model, sr41_X_train, sr41_y_train)
print(f"Naive Bayes runtime: {nb_time:.4f} seconds")

# Random Forest
rf_model = RandomForestClassifier(random_state=55041)
rf_time = time_model_run(rf_model, sr41_X_train, sr41_y_train)
print(f"Random Forest runtime: {rf_time:.4f} seconds")

# Extreme Gradient Boosting (XGBoost)
xgb_model = XGBClassifier(random_state=55041)
xgb_time = time_model_run(xgb_model, sr41_X_train, sr41_y_train)
print(f"XGBoost runtime: {xgb_time:.4f} seconds")
```

---

### Observations:

- **Fastest:** KNN and Naive Bayes are the fastest, taking **0.0051 seconds** and **0.0040 seconds**, respectively.  
- **Slowest:** Random Forest is the slowest, taking **1.0970 seconds**.  
- **Moderate:** Logistic Regression, SGD, and Decision Tree have moderate runtimes, ranging from **0.0170 seconds** to **0.0320 seconds**.  
- **Relatively Slow:** SVM and XGBoost are relatively slower, with runtimes of **0.5470 seconds** and **0.2892 seconds**, respectively.  

---

### Insights:

- **Fastest Models:**  
  - KNN and Naive Bayes are ideal for quick analysis or when computational resources are limited.  

- **Balanced Performance:**  
  - Logistic Regression, SGD, and Decision Tree offer a good trade-off between speed and complexity, making them suitable for general applications.  

- **Slowest Model:**  
  - Random Forest, while potentially more accurate, is the slowest. It may not be ideal for tasks requiring fast predictions.  

- **Relatively Slower Models:**  
  - SVM and XGBoost, with higher runtimes, might be more suitable for tasks where accuracy is critical, and computational resources are not a limiting factor.

##Memory Utilized


```python
import tracemalloc
from sklearn.linear_model import LogisticRegression, SGDClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier

def get_model_memory_usage(model, X, y):
    tracemalloc.start()
    model.fit(X, y)
    current, peak = tracemalloc.get_traced_memory()
    tracemalloc.stop()
    return peak / (1024 * 1024)  # Convert bytes to MB

# Logistic Regression
log_reg_model = LogisticRegression(random_state=55041)
log_reg_memory = get_model_memory_usage(log_reg_model, sr41_X_train, sr41_y_train)
print(f"Logistic Regression memory usage: {log_reg_memory:.2f} MB")

# Support Vector Machine (SVM)
svm_model = SVC(random_state=55041)
svm_memory = get_model_memory_usage(svm_model, sr41_X_train, sr41_y_train)
print(f"SVM memory usage: {svm_memory:.2f} MB")

# Stochastic Gradient Descent (SGD)
sgd_model = SGDClassifier(random_state=55041)
sgd_memory = get_model_memory_usage(sgd_model, sr41_X_train, sr41_y_train)
print(f"SGD memory usage: {sgd_memory:.2f} MB")

# Decision Tree
dt_model = DecisionTreeClassifier(random_state=55041)
dt_memory = get_model_memory_usage(dt_model, sr41_X_train, sr41_y_train)
print(f"Decision Tree memory usage: {dt_memory:.2f} MB")

# K-Nearest Neighbors (KNN)
knn_model = KNeighborsClassifier(n_neighbors=5)
knn_memory = get_model_memory_usage(knn_model, sr41_X_train, sr41_y_train)
print(f"KNN memory usage: {knn_memory:.2f} MB")

# Naive Bayes
nb_model = GaussianNB()
nb_memory = get_model_memory_usage(nb_model, sr41_X_train, sr41_y_train)
print(f"Naive Bayes memory usage: {nb_memory:.2f} MB")

# Random Forest
rf_model = RandomForestClassifier(random_state=55041)
rf_memory = get_model_memory_usage(rf_model, sr41_X_train, sr41_y_train)
print(f"Random Forest memory usage: {rf_memory:.2f} MB")

# Extreme Gradient Boosting (XGBoost)
xgb_model = XGBClassifier(random_state=55041)
xgb_memory = get_model_memory_usage(xgb_model, sr41_X_train, sr41_y_train)
print(f"XGBoost memory usage: {xgb_memory:.2f} MB")
```

### Observations:

- **Lowest:** Naive Bayes has the lowest memory usage, requiring only **0.11 MB**.  
- **Moderate:** Logistic Regression, Decision Tree, and KNN have moderate memory usage, ranging from **0.22 MB to 0.34 MB**.  
- **Relatively Low:** SVM and SGD have relatively low memory usage, requiring **0.26 MB** and **0.17 MB**, respectively.

### Insights:
